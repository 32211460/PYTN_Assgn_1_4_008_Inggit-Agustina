# Preprocessing

Mengimport library:

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

Membaca data:

```python
df = pd.read_csv('london_crime_by_lsoa.csv')
```

Menghilangkan kolom yang tidak diperlukan

```python
df.drop(['lsoa_code', 'borough'], axis=1, inplace=True)
```

Mengganti nilai yang hilang (NaN) pada kolom major_category dan minor_category dengan nilai mode:

```python
df['major_category'].fillna(df['major_category'].mode()[0], inplace=True)
df['minor_category'].fillna(df['minor_category'].mode()[0], inplace=True)
```

Mengganti nilai string pada kolom month menjadi tipe datetime:

```python
df['month'] = pd.to_datetime(df['month'])
```

Mengganti nilai kategorikal pada kolom major_category dan minor_category menjadi nilai numerik:

```python
le = LabelEncoder()
df['major_category'] = le.fit_transform(df['major_category'])
df['minor_category'] = le.fit_transform(df['minor_category'])
```

 Menghapus baris yang mengandung nilai NaN:

```python
df.dropna(inplace=True)
```

Mengganti tipe data kolom value menjadi integer:

```python
df['value'] = df['value'].astype(int)
```

 Menampilkan 5 baris pertama dari data setelah preprocessing:

```python
print(df.head())
```

# Membuat Area Plot

Mengimport library:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Membaca Data:

```python
df = pd.read_csv("london_crime_by_lsoa.csv")
```

Preprocessing:

```python
df = df[df['year'] == 2016]  # Filter data only for year 2016
df = df.groupby('borough')['value'].sum().reset_index()  # Aggregate total crime value by borough
```

Membuat Area plot:

```python
plt.figure(figsize=(12, 6))
plt.fill_between(df['borough'], df['value'], alpha=0.5)
plt.plot(df['borough'], df['value'], alpha=0.8, color='black')
plt.xticks(rotation=90)
plt.xlabel('Borough')
plt.ylabel('Total Crime')
plt.title('London Crime Analysis - Total Crime by Borough in 2016')
plt.show()
```



Borough dengan area plot yang lebih tinggi menunjukkan jumlah kejahatan total yang lebih tinggi, sementara borough dengan area plot yang lebih rendah menunjukkan jumlah kejahatan total yang lebih rendah. Hal ini dapat memberikan gambaran visual tentang perbandingan tingkat kejahatan antara berbagai borough di London.

Penggunaan area plot dapat membantu dalam mengidentifikasi perubahan tren jumlah kejahatan total di berbagai borough dari waktu ke waktu. Jika terdapat perubahan dalam luas area plot dari tahun ke tahun, maka hal ini dapat menjadi indikator perubahan jumlah kejahatan total di suatu borough.

Area plot juga dapat digunakan untuk membandingkan jumlah kejahatan total antara beberapa kategori kejahatan berbeda, misalnya berdasarkan jenis kejahatan atau wilayah geografis lainnya.

Penggunaan rotasi label pada sumbu x-axis dapat memudahkan pembaca dalam membaca dan menginterpretasi informasi pada plot, terutama jika terdapat banyak kategori yang ditampilkan.

# Membuat Bar Chart

Mengimport library:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

 Memfilter data bedasarkan tahun 2016:

```python
df = df[df['year'] == 2016]
```

 Mengelompokkan data dengan borough:

```python
boroughs = df.groupby('borough')['value'].sum().reset_index()
```

Mengurutkan data bedasarkan nilai dalam urutan menurun:

```python
boroughs = boroughs.sort_values(by='value', ascending=False)
```

Membuat bar chart:

```python
plt.bar(boroughs['borough'], boroughs['value'])

# atur judul bagan dan label sumbu
plt.title('Total Reported Crimes by Borough in London (2016)')
plt.xlabel('Borough')
plt.ylabel('Total Reported Crimes')

# rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# display chart
plt.show()
```

# Membuat Box Plot

Mengimport library:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Membaca data kejahatan pada tahun 2016

```python
df = pd.read_csv('london_crime_by_lsoa.csv')
df = df[df['year'] == 2016]
```

Menyiapkan data untuk box plot

```python
data = [df[df['borough'] == 'Westminster']['value'], 
        df[df['borough'] == 'Lambeth']['value'], 
        df[df['borough'] == 'Camden']['value'], 
        df[df['borough'] == 'Southwark']['value']]
```

Menyiapkan label untuk sumbu x

```python
labels = ['Westminster', 'Lambeth', 'Camden', 'Southwark']
```

Membuat Box Plot

```python
plt.boxplot(data, labels=labels)

# Menambahkan label pada sumbu x dan y
plt.xlabel('Borough')
plt.ylabel('Crime count')

# Menambahkan judul
plt.title('Box Plot of London Crime in 2016')

# Menampilkan plot
plt.show()
```

# Membuat Word Cloud

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv("london_crime_by_lsoa.csv")

# Menggabungkan semua kolom 'Major Category' menjadi satu string
text = " ".join(df['major_category'])

# Membuat objek WordCloud
wordcloud = WordCloud(width=800, height=800,
                      background_color='white',
                      min_font_size=10).generate(text)

# Menampilkan gambar WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()





```

# Membuat Folium Map

```python
import pandas as pd
import folium

# Membaca dataset
df = pd.read_csv('london_crime_by_lsoa.csv')

# Mengambil data kejahatan pada tahun 2020
df = df[df['year'] == 2020]

# Mengambil data koordinat dari file excel yang sudah disiapkan
coord_df = pd.read_csv('london_crime_by_lsoa.csv')

# Menggabungkan dataset kejahatan dengan dataset koordinat
df = df.merge(coord_df, on='lsoa_code')

# Menginisialisasi peta
m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)

# Menambahkan mark pada lokasi kejahatan
for index, row in df.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], 
                  popup=row['major_category'], 
                  icon=folium.Icon(color='red', icon='info-sign')).add_to(m)

# Menampilkan peta
m
```

# Overall Analysis

```python
Mengimport library:
```

```python
import matplotlib.pyplot as plt
```

Data analisis kejahatan di London:

```python
tahun = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
narkoba = [15339, 17137, 19390, 24586, 27439, 28234, 32061, 33752, 30972, 24795]
pencurian = [55624, 51714, 48576, 46518, 44247, 46450, 46149, 44598, 41309, 33826]
kekerasan = [14162, 13837, 12932, 12291, 12290, 13128, 14167, 15000, 16838, 15768]
```

Membuat area plot:

```python
plt.stackplot(tahun, narkoba, pencurian, kekerasan,
              labels=['Narkoba', 'Pencurian', 'Kekerasan'],
              colors=['#ff9999', '#66b3ff', '#99ff99'],
              alpha=0.8)  # Mengatur warna dan transparansi

# Menambahkan legend
plt.legend(loc='upper left')

# Mengatur label sumbu x dan y
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kejahatan')

# Mengatur judul plot
plt.title('Analisis Kejahatan di London')

# Menampilkan plot
plt.show()

```

Hipotesis 1: Jumlah kasus narkoba di London cenderung meningkat sepanjang tahun 2011-2020, yang dapat dipengaruhi oleh faktor seperti peningkatan penggunaan narkoba, peningkatan penegakan hukum, atau perubahan dalam pelaporan data.

Hipotesis 2: Jumlah kasus pencurian di London menunjukkan tren penurunan sepanjang tahun 2011-2020, yang dapat dipengaruhi oleh faktor seperti peningkatan kebijakan keamanan, perbaikan infrastruktur, atau perubahan perilaku masyarakat dalam menghindari pencurian.

Hipotesis 3: Jumlah kasus kekerasan di London mengalami fluktuasi dalam periode 2011-2020, yang dapat dipengaruhi oleh faktor-faktor sosial, ekonomi, atau politik, seperti perubahan dalam dinamika sosial, krisis ekonomi, atau kebijakan keamanan.
